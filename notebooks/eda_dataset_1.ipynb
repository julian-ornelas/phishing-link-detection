#### Data Manipulation

```python
# Dropping both ID, and some features that require webpage parsing

df = df.drop(['id'], axis = 1)
df = df.drop([
    "PctExtHyperlinks", "PctExtResourceUrls", "ExtFavicon", "InsecureForms",
    "RelativeFormAction", "ExtFormAction", "AbnormalFormAction",
    "PctNullSelfRedirectHyperlinks", "FrequentDomainNameMismatch",
    "FakeLinkInStatusBar", "RightClickDisabled", "PopUpWindow",
    "SubmitInfoToEmail", "IframeOrFrame", "MissingTitle", "ImagesOnlyInForm",
    "PctExtResourceUrlsRT", "AbnormalExtFormActionR", "ExtMetaScriptLinkRT",
    "PctExtNullSelfRedirectHyperlinksRT"], axis=1)

# Check for missing values

df.isna().sum()
```

#### Data Visualization

```python
# Goes through all features and outputs graph

for col in df.select_dtypes(include=["int", "float"]).columns:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Checking distribution of output variable
sns.countplot(x=df["CLASS_LABEL"], palette='pastel', legend=False)
plt.title("Class Distribution")
plt.show()
```

Outlier Detection + Normalization

```python
for col in df.select_dtypes(include=["int", "float"]).columns:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# We are going to robust scale the features that look highly skewed.
from sklearn.preprocessing import RobustScaler

features_to_scale = [
    "SubdomainLevel", 
    "PathLevel", 
    "UrlLength", 
    "NumDash", 
    "NumDashInHostname", 
    "NumUnderscore", 
    "NumPercent", 
    "NumQueryComponents", 
    "NumAmpersand", 
    "NumHash", 
    "NumNumericChars", 
    "HttpsInHostname", 
    "HostnameLength", 
    "PathLength", 
    "QueryLength", 
    "NumSensitiveWords"
]

scaler = RobustScaler()

df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

#### Feature Importance

```python
# Computes MI score and outputs the score

from sklearn.feature_selection import mutual_info_classif

X = df.drop(columns=["CLASS_LABEL"])
y = df["CLASS_LABEL"]

mi_scores = mutual_info_classif(X, y, discrete_features="auto")

mi_scores_df = pd.DataFrame({"Feature": X.columns, "Mutual Information": mi_scores})
mi_scores_df.sort_values(by="Mutual Information", ascending=False)
```

```python
# Keep features with MI score >= 0.01

selected_features = mi_scores_df[mi_scores_df["Mutual Information"] >= 0.01]["Feature"].tolist()
df = df[selected_features + ["CLASS_LABEL"]]

df.info()
df.head()
```
